```python
import importlib
import sys

if 'src.config' in sys.modules:
    del sys.modules['src.config']

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format', '{:.2f}'.format)

sys.path.append('../')
from src.config import *

print(f"TRAIN_FILE = {TRAIN_FILE}")  
```

```python
data=pd.read_csv(TRAIN_FILE)
data.head()
```

Target Distribution 


```python
fig,axes=plt.subplots(1,2,figsize=(12,4))
data['TARGET'].value_counts().plot(kind='bar',ax=axes[0],color=['steelblue', 'tomato'])
axes[0].set_title('Distribution of Target Variable')
axes[0].set_xticklabels(['Repaid (0)', 'Defaulted (1)'], rotation=0)
#perentage distribution
sizes = data['TARGET'].value_counts(normalize=True) * 100
axes[1].pie(sizes, labels=['Repaid (0)', 'Defaulted (1)'], autopct='%1.1f%%', colors=['steelblue', 'tomato'])
axes[1].set_title('Percentage Distribution of Target Variable')

plt.suptitle('Class Imbalance Overview', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/target_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nDefault rate: {data['TARGET'].mean()*100:.2f}%")
print(f"Class imbalance ratio: 1 : {(1-data['TARGET'].mean())/data['TARGET'].mean():.0f}")
```

Missing Values 

```python
msn_val=data.isnull().sum()
msn_val_pct=msn_val/len(data)*100
msn_df=pd.DataFrame(
    {'Missing_count': msn_val,
    'Missing_Percentage': msn_val_pct
    }
).sort_values(by='Missing_Percentage', ascending=False)
msn_df=msn_df[msn_df['Missing_count']>0]
print(msn_df.head(30))

plt.figure(figsize=(12, 8))
msn_df.head(30)['Missing_Percentage'].plot(kind='barh', color='tomato')
plt.title('Top 30 Columns by Missing Value %', fontsize=14, fontweight='bold')
plt.xlabel('Missing %')
plt.axvline(x=40, color='black', linestyle='--', label='40% threshold')
plt.legend()
plt.tight_layout()
plt.savefig('../reports/missing_values.png', dpi=150, bbox_inches='tight')
plt.show()

#drop columns with more than 40% missing values
cols_to_drop = msn_df[msn_df['Missing_Percentage'] > 40].index.tolist()
print(f"\nColumns with >40% missing (candidates to drop): {len(cols_to_drop)}")
print(cols_to_drop)
```

DataTypes Overview

```python
numeric_col=data.select_dtypes(include=[np.number]).columns.tolist()
numeric_col=[c for c in numeric_col if c not in [ID_COL, "TARGET"]]
print(f"\nNumeric columns: {len(numeric_col)}")
cat_col=data.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns: {len(cat_col)}")
```

Feature distribution

```python
key_features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
fig, axes = plt.subplots(2,4,figsize=(18,8))
axes = axes.flatten()
for i,col in enumerate(key_features):
    if col in data.columns:
        data[data['TARGET']==0][col].dropna().plot(kind='hist', bins=50, ax=axes[i], alpha=0.6, color='steelblue', label='Repaid', density=True)
        data[data['TARGET']==1][col].dropna().plot(kind='hist',bins=50,ax=axes[i],color='tomato', label='Defaulted', density=True, alpha=0.7)
        axes[i].set_title(col,fontsize=12, fontweight='bold')
        axes[i].legend(fontsize=10)
plt.suptitle('Distribution of Key Numeric Features by Target Class', fontsize=16, fontweight='bold')
plt.savefig('../reports/key_numeric_distributions.png', dpi=150, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

Correlations

```python
corr=data[numeric_col+ ['TARGET']].corr()['TARGET'].drop('TARGET',axis=0)
corr=corr.abs().sort_values(ascending=False)
print(corr.head(10))

plt.figure(figsize=(10,6))
corr.head(30).plot(kind='barh', color='tomato')
plt.title('Correlation of Numeric Features with Target', fontsize=14, fontweight='bold')
plt.xlabel('Absolute Correlation with Target')
plt.tight_layout()
plt.savefig('../reports/correlation_with_target.png', dpi=150, bbox_inches='tight')
plt.show()
```
